Dieses Notebook ruft `main.py` aus dem Projekt auf und ist für Google Colab gedacht.

Ablauf:
1. Repo bereitstellen (z. B. klonen oder in `/content` hochladen).

3. `main.py` mit gewünschten Parametern ausführen.


```python
import os
import pathlib

# Optional: Repo klonen, falls es noch nicht vorhanden ist
!git clone https://github.com/TimBrielmayer/16_03_Studienarbeit projekt
# %cd projekt

# Falls das Notebook nicht im Projektordner liegt, hier setzen:
# PROJECT_ROOT = '/content/projekt'
PROJECT_ROOT = None

def find_project_root():
    if PROJECT_ROOT:
        return pathlib.Path(PROJECT_ROOT).resolve()
    cwd = pathlib.Path.cwd().resolve()
    for p in [cwd, *cwd.parents]:
        if (p / 'main.py').exists():
            return p
    for p in pathlib.Path('/content').rglob('main.py'):
        return p.parent
    raise FileNotFoundError('main.py nicht gefunden. Repo klonen/hochladen und PROJECT_ROOT setzen.')

root = find_project_root()
os.chdir(root)
print('Project root:', root)
```
